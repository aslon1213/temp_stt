```python
from faster_whisper import WhisperModel
import os
```

```python
import time
```

```python
model = WhisperModel(
        "aslon1213/whisper-small-uz-with-uzbekvoice-ct2", device="cuda"
    )
```

```python
import time
text = ""
audio = 0
with open("tttt.mp3", "rb") as f:
    audio = f.read(-1)
print(type(audio))
```

```python
import io
audio = io.BytesIO(audio)
```

```python
segments, info = model.transcribe(audio =audio, language="uz")
start = time.time()
text_chunks = []
for segment in segments:
    text += segment.text
    print("Segment :", )
    text_chunks.append({
        "start":segment.start, 
        "end":segment.end, 
        "text":segment.text
    })
    print(f"Processing segment [{segment.start}: {segment.end}] : {segment.text}" )
print(f"Time to process full text: {time.time() - start} -- gpu id:{0}")
```
